```python
import os
from rembg import remove
from PIL import Image
import numpy as np
import cv2
from sklearn.preprocessing import StandardScaler

# Paths
input_folder = r'C:\Users\Uma\Desktop\LAMBTON\TERM_2\AI LAb\dermatologic_classification-main\dermatologic_classification-main'
output_folder = r'C:\Users\Uma\Desktop\LAMBTON\TERM_2\AI_Lab\dermatologic_classification-main\dermatologic_classification-main\data\processed_images'

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)
# Function to preprocess each image
def preprocess_image(file_path, output_path):
    # Step 1: Background Removal
    input_image = Image.open(file_path)
    image_no_bg = remove(input_image)

    # Convert to numpy array
    image_array = np.array(image_no_bg)

    # Step 2: Resize to a fixed size
    target_size = (128, 128)  # For example, 128x128
    resized_image = cv2.resize(image_array, target_size)

    # Step 3: Normalization (scaling pixel values)
    normalized_image = resized_image / 255.0

    # Step 4: Convert back to uint8 for saving
    processed_image = Image.fromarray((normalized_image * 255).astype(np.uint8))

    # Save the processed image
    processed_image.save(output_path)
    
    # Loop through each image in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            file_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, f'processed_{filename}')
            preprocess_image(file_path, output_path)
    
```
